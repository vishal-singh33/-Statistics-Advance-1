**Question 1: What is a random variable in probability theory?**

answer-A random variable is a function that assigns a real number to each outcome in a sample space of a random experiment.

It allows us to quantify uncertain outcomes so that we can apply mathematical analysis.

There are two main types:

1.Discrete random variable - takes a countable number of values (e.g., number of heads in 3 coin tosses).

2.Continuous random variable - takes values from an interval of real numbers (e.g., the time it takes for a bus to arrive).

Toss a coin: sample space = {Heads, Tails}.

Define a randon variable X:
X = 1 if head, X = 0 if tail
here X is a discrete random variable

```python
import random
import numpy as np

# Example 1: Discrete Random Variable (Coin Toss)
def coin_toss():
    return random.choice([0, 1])  # 0 = Tail, 1 = Head

# Simulate 10 coin tosses
coin_results = [coin_toss() for _ in range(10)]
print("Coin Toss Results (Discrete Random Variable):", coin_results)

```

**Question 4: What is a binomial distribution, and how is it used in probability?**

Answer:-- Binomial Distribution

The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials, each with the same probability of success.

Conditions for a Binomial Experiment:

Fixed number of trials (n)
→ e.g., tossing a coin 10 times.

Two possible outcomes: Success (S) or Failure (F)
→ e.g., Head (success) or Tail (failure).

Constant probability (p) of success in each trial.
→ e.g., fair coin →
𝑝
=
0.5
p=0.5.

Independence of trials
→ outcome of one trial doesn’t affect another.

Uses of Binomial Distribution:

Modeling number of defective items in a batch.

Probability of X heads in coin tosses.

Pass/fail tests in quality control.

Customer buying/not buying a product.

in short-- The binomial distribution helps us calculate probabilities for scenarios with repeated, independent, yes/no type trials.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10   # number of trials (e.g., 10 coin tosses)
p = 0.5  # probability of success (e.g., probability of Head)

# 1. Compute probability of exactly k successes
k = 6
probability = binom.pmf(k, n, p)
print(f"Probability of getting exactly {k} successes out of {n} trials: {probability:.4f}")

# 2. Generate full distribution
x = np.arange(0, n+1)        # possible values: 0 to n
pmf = binom.pmf(x, n, p)     # probability mass function

# 3. Plot
plt.figure(figsize=(8, 5))
plt.stem(x, pmf, basefmt=" ")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.xlabel("Number of Successes (k)")
plt.ylabel("P(X = k)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Population: Exponential distribution (skewed, non-normal)
np.random.seed(42)
population = np.random.exponential(scale=2.0, size=100000)

# Function to simulate sampling means
def sample_means(sample_size, num_samples=1000):
    means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]
    return means

# Sample sizes to test
sample_sizes = [5, 30, 100]

# Plotting
plt.figure(figsize=(14, 8))

for i, n in enumerate(sample_sizes, 1):
    means = sample_means(n)
    plt.subplot(1, 3, i)
    plt.hist(means, bins=30, density=True, alpha=0.7, color="skyblue", edgecolor="black")
    plt.title(f"Sampling Distribution of Means\nSample Size n={n}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.grid(True)

plt.suptitle("Central Limit Theorem Demonstration\nPopulation: Exponential (Skewed)", fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

**Question 7: What is the significance of confidence intervals in statistical analysis?**

Answer:-- A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter (like mean or proportion) with a certain level of confidence.

Example:
If we compute a 95% confidence interval for the mean height of students and get
[
165
,
175
]
[165,175], we say:

“We are 95% confident that the true population mean lies between 165 and 175 cm.”

Why are Confidence Intervals Significant?

Quantify Uncertainty:
Instead of just giving a single estimate (like a sample mean), CI provides a range, showing the precision of the estimate.

More Informative than Point Estimates:

Point estimate: “Average income = $50,000.”

CI: “Average income is between $48,000 and $52,000 (95% CI).”
→ The CI shows the possible variation.

Connection to Hypothesis Testing:

If the CI for a mean difference excludes 0, it suggests a statistically significant difference.

Helps make decisions without relying only on p-values.

Reliability Measure:
A 95% CI doesn’t mean the parameter has a 95% chance of lying in that range (the parameter is fixed) — it means that if we repeated sampling many times, 95% of the calculated intervals would contain the true parameter.

Widely Used in Real Life:

Medical research (drug effectiveness ranges)

Economics (GDP growth estimates)

Polling (margin of error in surveys)

In short:
Confidence intervals are critical because they provide a range of plausible values for population parameters, quantify the uncertainty of estimates, and support statistical decision-making.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate sample data (e.g., student heights in cm)
np.random.seed(42)
data = np.random.normal(loc=170, scale=10, size=50)  # mean=170, std=10, sample size=50

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# 95% Confidence Interval using t-distribution
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower, ci_upper = sample_mean - margin_of_error, sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")

# Plot
plt.figure(figsize=(8, 5))
plt.hist(data, bins=10, color="skyblue", edgecolor="black", alpha=0.7, density=True)
plt.axvline(sample_mean, color="blue", linestyle="--", label=f"Sample Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color="red", linestyle="--", label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color="red", linestyle="--", label=f"95% CI Upper = {ci_upper:.2f}")

plt.title("95% Confidence Interval for Sample Mean")
plt.xlabel("Height (cm)")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as st

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to numpy array
data = np.array(daily_sales)

# Sample statistics
mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)
n = len(data)

# 95% Confidence Interval using t-distribution
alpha = 0.05
t_critical = st.t.ppf(1 - alpha/2, df=n-1)
margin_of_error = t_critical * (std_sales / np.sqrt(n))

ci_lower = mean_sales - margin_of_error
ci_upper = mean_sales + margin_of_error

print(f"Sample Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```
